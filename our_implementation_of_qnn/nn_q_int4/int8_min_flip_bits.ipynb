# Minimum flip bits for int8 attacks, (paper)

```python
import PyPDF2
import numpy as np
from scipy.spatial.distance import hamming
```

```python
def extract_words_from_pdf(file_path):
    pdf_file = open(file_path, 'rb')
    reader = PyPDF2.PdfReader(pdf_file)

    word_array = []
    for page in reader.pages:
        text = page.extract_text()
        words = text.split()  # Split text into words
        word_array.extend(words)  # Add words to the array

    pdf_file.close()
    return word_array
```

```python
def table_code_words(input_array):

    table_code_words = np.zeros((255,), dtype=np.ndarray)

    for i in range(255):
        table_code_words[i] = input_array[3 + i*4]

    return table_code_words
```

```python
def table_binary_code_words(input_array):

    table_code_words_binary = np.zeros((255,), dtype=np.ndarray)

    for i in range(255):
        if i < 127:
            table_code_words_binary[i + 128] = [int(bit) for bit in input_array[3 + i*4]]
        else:
            table_code_words_binary[i - 127] = [int(bit) for bit in input_array[3 + i*4]]

    return table_code_words_binary
```

```python
pdf_file_path = "D:\DP\DP\Code\HammingCode(12,8,3)_C12-3.pdf"
C_12_3 = extract_words_from_pdf(pdf_file_path)
```

```python
# Print the extracted text
for text in C_12_3:
    print(text)
```

```python
table_code_words_binary = table_binary_code_words(C_12_3)
print(table_code_words_binary)

```

```python
table_hamming_distance = np.full((255,255), 255, dtype=np.uint8)
np.fill_diagonal(table_hamming_distance, 0)
print(table_hamming_distance)
```

```python
for i in range(255):
    for j in range(255):
        if i != j:
            code_word_bin = table_code_words_binary[i]
            code_word_bin_2 = table_code_words_binary[j]
            table_hamming_distance[i][j] = hamming(code_word_bin, code_word_bin_2) * len(code_word_bin)

```
